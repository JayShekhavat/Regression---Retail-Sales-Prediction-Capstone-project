### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
sort_df = np.sort(df['Customers'])

Q1 = np.percentile(sort_df, 25, interpolation = 'midpoint')
Q2 = np.percentile(sort_df, 50, interpolation = 'midpoint')
Q3 = np.percentile(sort_df, 75, interpolation = 'midpoint')

print('Q1 25 percentile of the given data is, ', Q1)
print('Q2 50 percentile of the given data is, ', Q2)
print('Q3 75 percentile of the given data is, ', Q3)

IQR = Q3 - Q1
print('Interquartile range is', IQR)

````

````python
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
````

````python
outlier =[]
for x in sort_df:
    if ((x> up_lim) or (x<low_lim)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
````

````python

def removal_box_plot(df, column, threshold):
	sns.boxplot(df[column])
	plt.title(f'Original Box Plot of {column}')
	plt.show()

	removed_outliers = df[df[column] <= threshold]

	sns.boxplot(removed_outliers[column])
	plt.title(f'Box Plot without Outliers of {column}')
	plt.show()
	return removed_outliers


threshold_value = 1200

no_outliers = removal_box_plot(df, 'Customers', threshold_value)
````
